```python
import itertools
```

```python
DATA = """
1721
979
366
299
675
1456
""".strip()
```

```python
def product(values, n=2):
    """
    Returns all possible (ordered) subsets of the given size
    from the given values.
    
    :param values: list(int,)
    :param n: int
    :return: list(tuple(int,),)
    """
    buckets = (tuple(values) for _ in range(n))
    return itertools.product(*buckets)
```

```python
def prod(values):
    """
    Returns the product of the given values.
    
    :param values: list(int,)
    :return: int
    """
    return (values[0] * prod(values[1:])) if values else 1
```

```python
def find(numbers, target, n):
    """
    Finds which subset of the given size from the given numbers
    adds up to the given target and returns it and the product
    of the values in the subset.
    
    :param numbers: list(int,)
    :param target: int
    :param n: int
    :return: subset:tuple(int,), product:int
    """
    for someNumbers in product(numbers, n=n):
        if sum(someNumbers) == target:
            return someNumbers, prod(someNumbers)
```

```python
numbers = sorted(map(int, DATA.splitlines()))
numbers
```

```python
_, firstSolution = find(numbers, 2020, 2)
_, secondSolution = find(numbers, 2020, 3)
firstSolution, secondSolution
```
