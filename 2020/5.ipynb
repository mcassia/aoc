```python
def convert(x, d=7):
    a = x[:d].replace('F', '0').replace('B', '1')
    b = x[d:].replace('L', '0').replace('R', '1')
    row = int(a, 2)
    column = int(b, 2)
    return row * 8 + column
```

```python
def convert(x):
    rowRange = [0, 128]
    columnRange = [0, 8]
    
    for c in x[:7]:
        if c == 'F':
            rowRange[1] = rowRange[1] - (rowRange[1] - rowRange[0]) / 2
        else:
            rowRange[0] = rowRange[0] + (rowRange[1] - rowRange[0]) / 2

    for c in x[7:]:
        if c == 'L':
            columnRange[1] = columnRange[1] - (columnRange[1] - columnRange[0]) / 2
        else:
            columnRange[0] = columnRange[0] + (columnRange[1] - columnRange[0]) / 2
    
    return int(rowRange[0] * 8 + columnRange[0])
```

```python
ids = list(map(convert, DATA.splitlines()))
```

```python
for i in range(min(ids), max(ids)):
    if i not in ids:
        print(i)
```
